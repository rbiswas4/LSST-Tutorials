As an example, let's calculate some qualities of the Hour Angle distribution -- first, at each point over the sky (e.g. we'll calculate the full range of the hour angle at each point in the sky) and second, for all visits in each filter (e.g. we'll generate a histogram for all visits) -- and then demonstrate reading the metric data from disk back into memory.

Import the matplotlib and [MAF](https://confluence.lsstcorp.org/display/SIM/MAF+documentation) modules we will use.

```python
import matplotlib.pyplot as plt
%matplotlib inline

from lsst.sims.maf.db import OpsimDatabase
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.sliceMetrics as sliceMetrics
from lsst.sims.maf.utils import opsimUtils
```

Connect to an [opsim sqlite database](https://confluence.lsstcorp.org/display/SIM/OpSim+Datasets+for+Cadence+Workshop+LSST2015) which we have downloaded and saved on our local machine. (I usually just symlink it into my working directory, but you can provide a full path instead).

```python
runName = 'lucy_1002'
opsdb = OpsimDatabase('sqlite:///' + runName + '_sqlite.db')
```

----

Starting with the first example, calculating the full range of the hour angle distribution at each point in the sky.

Instantiate and set up my [metric](https://confluence.lsstcorp.org/display/SIM/MAF+Metrics) object to calculate the full range value of the HA across the sky. 

```python
metricList = []
metricList.append(metrics.FullRangeMetric(col='HA'))
```

Instantiate my slicer (using the default fieldRA/fieldDec values, so no dithering). Because we want to calculate a metric across the sky, we're using a [HealpixSlicer](https://confluence.lsstcorp.org/display/SIM/MAF+Slicers).

```python
slicer = slicers.HealpixSlicer(nside=64)
```

Instantiate the RunSliceMetric (which we'll use to run the metrics above over the slicer above). 

```python
outDir = 'ha_test'
sm = sliceMetrics.RunSliceMetric(outDir=outDir)
```

Set the metrics/slicer in the RunSliceMetric, so that we can determine the columns we need from the database. If we wanted to use a customized [stacker](https://confluence.lsstcorp.org/display/SIM/MAF+Stackers), we would instantiate that and pass it as well; since we're going to use the Hour Angle stacker in its default configuration, we can just let MAF handle it.

```python
sm.setMetricsSlicerStackers(metricList, slicer)
```

Let MAF find the data columns needed from the opsim database for these metrics, stackers and slicer.

```python
dbcols = sm.findDataCols()
# see? this is what we'll query for -- note that the columns that the HA stacker needs have been added automatically
print dbcols
```

Get the data from the sqlite database, subject to a sqlconstraint (something like 'filter="r"' or 'night<365'). We also pass the stackers from the RunSliceMetric, so the columns which will be generated by the stackers can be added to the simdata results, if needed. 

```python
sqlconstraint = 'filter="r"'  #no sql constraint
simdata = opsimUtils.getSimData(opsdb, sqlconstraint, dbcols, sm.stackerObjs)
# print the names of the simdata columns - note that the HA column has been calculated and added into simdata.
print simdata.dtype.names
```

Now that we have the data in hand, we have to set up our slicer. In this step, the slicer creates indexes in the simdata array so that it can subdivide the visits quickly when generating the metric data.

```python
sm.slicer.setupSlicer(simdata)
```

Calculate the metric values at all the 'slicePoints' (subdivisions) in the slicer. Pass along some metadata relevant to this simData set so that it can be incorporated into titles/labels/filenames.

```python
sm.runSlices(simdata, simDataName=runName, sqlconstraint=sqlconstraint)
```

Now our metric data has been generated! 

```python
# For those who want to know:
#  where their metric data is being stored and some of the metadata which is associated with each metric.
for iid in sm.metricValues:
    print iid, '__', sm.metricNames[iid], '__',  sm.metadatas[iid], '__', sm.slicers[iid].slicerName, sm.metricValues[iid]
```

Save the metric data to disk.

```python
iid = 0
sm.writeMetric(iid)
!ls $outDir
```

Generate all plots for this metric.

```python
sm.plotMetric(iid)
```

Generate some [summary statistics](https://confluence.lsstcorp.org/display/SIM/MAF+Summary+Statistics) -- for example, the median and RMS of the "Full Range HA" metric results. This consolidates our values at each point across the sky into a single value.

```python
median = sm.computeSummaryStatistics(iid, metrics.MedianMetric('metricvalue'))
rms = sm.computeSummaryStatistics(iid, metrics.RmsMetric('metricvalue'))
print 'Median:', median,  'RMS:', rms
```

----

Moving on to our second example, generating the histogram of the hour angle for all the (r band) visits in simdata (we'll just reuse the simdata array).

Instantiate our metrics - this time we will use a 'count' metric, as that's what you do when you generate a histogram. The column used doesn't really matter here, but using 'HA' means that we won't fetch data we weren't already going to get anyway.

```python
metricList = [metrics.CountMetric(col='HA')]
```

Instantiate our slicer - to generate a histogram, we'll use a OneDSlicer, subdividing on HA.

```python
slicer = slicers.OneDSlicer(sliceColName='HA', binsize=0.25)
```

And like before, instantiate the RunSliceMetric and set the metrics/slicers/stackers.

```python
sm = sliceMetrics.RunSliceMetric(outDir=outDir)
sm.setMetricsSlicerStackers(metricList, slicer)
```

We're re-using our old simdata array, because we know that we have all the data & columns we need, so we can skip getting the column names and the simdata array. 

Set up the indexing in the slicer.

```python
sm.slicer.setupSlicer(simdata)
```

Generate the metric data. 

```python
sm.runSlices(simdata, simDataName=runName, sqlconstraint=sqlconstraint, metadata='r band')
```

Save it to disk. Note that now we have a slightly different filename (and different plot labels below) because we set 'metadata' above. The RunSliceMetric uses that information if available, instead of the sqlconstraint.

```python
iid = 0
sm.writeMetric(iid)
```

Generate plots.

```python
sm.plotMetric(iid)
```

If I want the median hour angle of all visits, I can calculate that too - but not directly from the above metric data. The metric data calculated above is the histogram .. counts of the number of visits at each hour angle bin. The way to calculate the median (or mean or rms ...) hour angle of all visits is to use the 'unislicer', which simply uses all visits to calculate this quantity directly. 

```python
metric = metrics.MeanMetric(col='HA')
slicer = slicers.UniSlicer()
sm = sliceMetrics.RunSliceMetric(outDir=outDir)
sm.setMetricsSlicerStackers(metric, slicer)
slicer.setupSlicer(simdata)
sm.runSlices(simdata, simDataName=runName, sqlconstraint=sqlconstraint, metadata='r band')
sm.writeAll()
```

You can then either use the metric value directly, or compute a summary statistic using the "IdentityMetric". The advantage of using the summary statistic is that then this information is preserved in the resultsDb_sqlite.db file inside your output directory - all summary statistics are saved into a table in this sqlite file, which can be useful later.

```python
print 'Just the metric value', sm.metricValues[0]
print 'Summary statistic', sm.computeSummaryStatistics(0, metrics.IdentityMetric(col='metricvalues'))
```

It often happens that you want to replot metric data or use it in combination with other metric data to do something else. So it's useful to know how to read data back from disk into a sliceMetric. MAF stores data on disk in special npz files. These files retain information on the slicer properties as well as the metric values.

```python
filenames = !ls $outDir/*npz 
print filenames
```

```python
sm = sliceMetrics.BaseSliceMetric(useResultsDb=False, outDir=outDir)
```

```python
iids = sm.readMetricData(filenames)
```

```python
# As an illustration of where that data went / how it can be accessed if needed.
for iid in iids:
    print iid, sm.metricNames[iid], sm.slicers[iid].slicerName, sm.metricValues[iid].mean()
```

Let's do some tweaks to the FullRange HA plot - we'll use the [plotDict](https://confluence.lsstcorp.org/display/SIM/MAF+Metrics) plot parameters to set the limits to be 0/15. (although, alas, there is only so much forcing you can do to matplotlib .. so let's focus on making it look nicer by changing the color bar format too).

```python
iid = 1
# Change ranges for the histogram
sm.plotDicts[iid]['xMin'] = 0
sm.plotDicts[iid]['xMax'] = 15
# Change ranges for the sky map
sm.plotDicts[iid]['colorMin'] = 0
sm.plotDicts[iid]['colorMax'] = 15 
# Change the colorbar/xaxis format for both
sm.plotDicts[iid]['cbarFormat'] = '%d'
```

```python
sm.plotMetric(iid)
```
