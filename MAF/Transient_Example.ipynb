#Example Running Transient Metric#

The motivation is that there are a large number of transient astronomical objects that can be approximated with a simple rising and falling light curve.  This metric computes what fraction of transients would be detected if the transients were continuously happening.  For example, if a transient has a 10-day duration, the metric assumes 365 back-to-back transients over a 10-year survey and then computes what fraction of these meet the detection criteria.

The metric can be configured to use various thresholds before counting a light curve as "observed".  

```python
%matplotlib inline
```

```python
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
from lsst.sims.maf.sliceMetrics import RunSliceMetric
from lsst.sims.maf.db import OpsimDatabase
from lsst.sims.maf.utils import opsimUtils
```

Set up a healpix slicer to use the dithered RA and dec positions

```python
nside = 32
slicer = slicers.HealpixSlicer(nside=nside, spatialkey1="ditheredRA", spatialkey2="ditheredDec")
```

Configure a few transient metrics.

First, all the transients have the properties that they are 60 days in duration, peaking after 10 days.  The light curves linearly rise with a slope of -0.5 mag/day, and decline with 0.1 mag/day (in all filters, so there is no color evolution).  The peak brightness is r=20 (which is the default for all fitlers).

For the first metric, we set no detection criteria, so if the light curve is observed above the 5-sigma threshold in any filter, it is counted as detected.

For the second metric, we demand that there be two observations before the light curve peaks.  There is also the constraint that there must be 5 sections of the light curve observed in the same filter.  

Finally, we demand 5 sections of the light curve be observed in 2 unique filters.

```python
metricList = []
transProperties = {'riseSlope':-0.5, "declineSlope":0.1, "transDuration":60., 'peakTime':10., 'rPeak':20.}
# No criteria, any light curve that is detected in any filter is counted
detectCriteria = {}
kwargs = transProperties.copy()
kwargs['metricName'] = 'Detected at all'
metricList.append(metrics.TransientMetric(**kwargs) )
# Demand 2 points on the rise, and 3 points spread out accross the LC
kwargs = transProperties.copy()
kwargs['metricName'] = 'Well Sampled'
detectCriteria = {"nPrePeak":2, "nPerLC":5}
kwargs.update(detectCriteria)
metricList.append(metrics.TransientMetric(**kwargs) )
# Same as before, but now in at least 2 filters
detectCriteria = {"nPrePeak":2, "nPerLC":5, "nFilters":2}
kwargs = transProperties.copy()
kwargs['metricName'] = 'Well Sampled 2 Filters'
kwargs.update(detectCriteria)
metricList.append(metrics.TransientMetric(**kwargs) )
```

```python
rsm = RunSliceMetric(outDir='TransOut')
rsm.setMetricsSlicerStackers(metricList, slicer)
```

```python
dbcols = rsm.findDataCols()
print dbcols
```

```python
runName = 'enigma_1189'
opsdb = OpsimDatabase('sqlite:///' + runName + '_sqlite.db')
```

```python
# Grab the data. This should get simpler soon.
sqlconstraint = '' # All the data
simdata = opsimUtils.getSimData(opsdb, sqlconstraint, dbcols, rsm.stackerObjs)
```

```python
rsm.runSlices(simdata, simDataName=runName, sqlconstraint=sqlconstraint)
```

Setting many detection criteria can be relatively expensive in Python, since it involves running a nested for-loop with multiple if-statements.  We've made this as efficient as possible using the np.searchsorted function.  The hope is that this metric will serve as a good starting point for the transient community to build metrics for their particular objects of interest.
